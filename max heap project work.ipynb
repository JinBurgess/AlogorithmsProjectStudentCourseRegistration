```python
def Swap(list, pos1, pos2):

    list[pos1], list[pos2] = list[pos2], list[pos1]

    return list

def buildmaxheap(dictionary,list):
    for index in range(len(list)-1,-1,-1):
        maxheapify(dictionary,list,index)
    return(list)

def maxheapify(dictionary,list,index):
    L_child = left(index)
    R_child = right(index)
    if L_child <= heapsize(list) and dictionary[list[L_child]]>dictionary[list[index]]:
        largest = L_child
    else:
        largest = index
    # end if 
    if R_child<=heapsize(list) and dictionary[list[R_child]]>dictionary[list[largest]]:
        largest = R_child
    # end if
    if largest != index:
        list = Swap(list,index,largest) 
        return(maxheapify(dictionary,list,largest))

def heapsize(list):
    return(len(list)-1)

def parent(index):
    return((index-1)//2)

def left(index):
    return((2*index)+1)

def right(index):
    return((2*index)+2)

class algorithmsregistration:
    def __init__(self, maxstudents):
        self.maxstudents = maxstudents
        self.level_dictionary = {}
    def register_student(self,name,year,registrationday,major): 
        level = 0 #reset the level for each function call
        if year == 0: #for auditors
            level = level + (.5-(.1*(registrationday))) # level will equal 0.5 for registrations before orientation week, 0.4 for registrations on the first day, etc.
            self.level_dictionary.update(
            {name:level}
            )
            return(self.level_dictionary) #returns the updated dictionary
        if year <5: # for undergraduate students
            if major == 'cs' or major == 'math':
                level = level+10 # adds 10 to the level for cs or math majors,
            
            level = level + year # adds 1,2,3, or 4 to the level based on the year. 
            level = level +(.5-(.1*(registrationday))) #+0.5 for students registered before the orientation week, same as for auditors.
        else: level = level +15 #for graduate students, level will automatically be 15.
        
        self.level_dictionary.update(
            {name:level}
        )
        return(self.level_dictionary) #return the updated dictionary

    def unenroll_student(self,name):
        del self.level_dictionary[name]
        return (self.level_dictionary)
    
    def enrollment_list(self):
        self.enrollment_key_list = []
        self.enrolled_students = []
        self.waitlisted_students = []
        self.heapsize = heapsize(self.enrollment_key_list)
        for key in self.level_dictionary.keys():
            self.enrollment_key_list.append(key)
        buildmaxheap(self.level_dictionary,self.enrollment_key_list)
        for index in range(len(self.enrollment_key_list)-1,1,-1):
            Swap(self.enrollment_key_list,1,index)
            self.heapsize = self.heapsize-1
            maxheapify(self.level_dictionary,self.enrollment_key_list,index)
        
        if len(self.enrollment_key_list)>self.maxstudents:
            self.enrolled_students = self.enrollment_key_list[0:(self.maxstudents-1)]
            self.waitlisted_students = self.enrollment_key_list[self.maxstudents:]
            return self.enrolled_students and self.waitlisted_students
        
        else: self.enrolled_students = self.enrollment_key_list
        return self.enrolled_students



    def maxheapify(self,list,index):
        L_child = left(index)
        R_child = right(index)
        if L_child <= heapsize(list) and self.level_dictionary[list[L_child]]>self.level_dictionary[list[index]]: #this function is slightly different than the previous because of the use of the dictionary
            largest = L_child
        else:
            largest = index
        # end if 
        if R_child<=heapsize(list) and self.level_dictionary[list[R_child]]>self.level_dictionary[list[largest]]: # also different here. 
            largest = R_child
        # end if
        if largest != index:
            list = Swap(list,index,largest) 
            return(self.maxheapify(list,largest))
        

    def buildmaxheap(self):
        self.level_key_list =[] # this will create a list of the key values to use to build the max heap.
        for key in self.level_dictionary.keys():
            self.level_key_list.append(key)
        for index in range(len(self.level_key_list)-1,-1,-1):
            self.maxheapify(self.level_key_list,index)
        return(self.level_key_list)
        #for
```
